```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
data=pd.read_csv("mall_customer.csv")
data
X,y = make_blobs(n_samples = 200,n_features = 2,centers = 3,random_state = 23)

fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
k = 3

clusters = {}
np.random.seed(23)

for idx in range(k):
    center = 2*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center' : center,
        'points' : []
    }
    
    clusters[idx] = cluster
    
clusters
```

```python
plt.scatter(X[:,0],X[:,1])
plt.grid(True)
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.show()
```

```python
def distance(p1,p2):
    return np.sqrt(np.sum((p1-p2)**2))
def assign_clusters(X, clusters):
    for idx in range(X.shape[0]):
        dist = []
        
        curr_x = X[idx]
        
        for i in range(k):
            dis = distance(curr_x,clusters[i]['center'])
            dist.append(dis)
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)
    return clusters

def update_clusters(X, clusters):
    for i in range(k):
        points = np.array(clusters[i]['points'])
        if points.shape[0] > 0:
            new_center = points.mean(axis =0)
            clusters[i]['center'] = new_center
            
            clusters[i]['points'] = []
    return clusters
def pred_cluster(X, clusters):
    pred = []
    for i in range(X.shape[0]):
        dist = []
        for j in range(k):
            dist.append(distance(X[i],clusters[j]['center']))
        pred.append(np.argmin(dist))
    return pred 
clusters = assign_clusters(X,clusters)
clusters = update_clusters(X,clusters)
pred = pred_cluster(X,clusters)
```

```python
plt.scatter(X[:,0],X[:,1],c = pred)
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
X, y = load_iris(return_X_y=True)
```

```python
sse = [] #SUM OF SQUARED ERROR
for k in range(1,11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11), y=sse)

g.set(xlabel ="Number of cluster (k)", 
      ylabel = "Sum Squared Error", 
      title ='Elbow Method')

plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
pred = kmeans.fit_predict(X)
pred
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[:2]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
    
plt.subplot(1,2,2)   
plt.scatter(X[:,2],X[:,3],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[2:4]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd

# Example data - Replace X with your dataset
X = np.random.rand(100, 2)
pca = PCA(n_components=2)
X_principal = pd.DataFrame(pca.fit_transform(X), columns=['P1', 'P2'])

# Run DBSCAN
db = DBSCAN(eps=0.0375, min_samples=50).fit(X_principal)
labels = db.labels_

# Define colors for each label
colours1 = {0: 'r', 1: 'g', 2: 'b', 3: 'c', 4: 'y', 5: 'm', -1: 'k'}
cvec = [colours1[label] if label in colours1 else 'k' for label in labels]

# Plotting
plt.figure(figsize=(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c=cvec)

# Create a legend
r = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='r', markersize=10)
g = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='g', markersize=10)
b = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='b', markersize=10)
c = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='c', markersize=10)
y = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='y', markersize=10)
m = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='m', markersize=10)
k = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='k', markersize=10)

plt.legend([r, g, b, c, y, m, k],
           ['Label 0', 'Label 1', 'Label 2', 'Label 3', 'Label 4', 'Label 5', 'Label -1'],
           scatterpoints=1, loc='upper left', ncol=3, fontsize=8)

plt.show()
```
